```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
train  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/startupdata/training_set_label.csv" )
test = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/startupdata/testing_set_label.csv')
```

```
train.shape,test.shape
```

```
train.head()
```

```
train=train.drop(['Unnamed: 0','id','Unnamed: 6','name'],axis=1)
test=test.drop(['Unnamed: 0','id','Unnamed: 6','name'],axis=1)
```

```
combine=train.append(test)
combine.reset_index(inplace=True)
combine.head()
```

```
closedat=[]
for i in combine['closed_at']:
  if len(str(i))==3:
    closedat.append(0)
  else:
    closedat.append(1)
combine['is_closed']=closedat
```

```
combine.head()
```

```
combine.drop(['state_code.1','index'],axis=1,inplace=True)
```
